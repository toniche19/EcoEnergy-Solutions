**`## 4.1 Verificación de valores fuera de rango`**

En este apartado se comprueba si existen valores fuera de los rangos lógicos esperados en las variables numéricas del conjunto de datos.

```python
# VERIFICACION DE VALORES FUERA DE RANGO
#==================================================
# 4.1 Verificacion de valores fuera de rango
# =================================================

# Comprobamos si existen valores negativos en las variables numéricas

print("Consumo negativo:", (df["consumption_kwh"] < 0).sum())
print("Emisiones negativas:", (df["co2_emissions"] < 0).sum())
print("Coste por kWh negativo:", (df["cost_per_kwh"] < 0).sum())
print("Coste total negativo:", (df["total_cost"] < 0).sum())

# Comprobamos si el porcentaje de energía renovable está fuera del rango 0-100

fuera_rango = ((df["renewable_energy_percentage"] < 0) |
               (df["renewable_energy_percentage"] > 100)).sum()

print("Porcentaje de energía renovable fuera del rango (0-100):", fuera_rango)
```

**`## 4.2 Comprobación de consistencia`**

En este apartado se verifica que el coste total coincide con el resultado de multiplicar el consumo energético por el coste por kWh.

```python
# COMPROBACION DE CONSISTENCIA DE LOS DATOS
#=================================================
# 4.2 Comprobacion de consistencia
# ============================================

# Calculamos el coste total esperado
coste_calculado = df["consumption_kwh"] * df["cost_per_kwh"]

# Comparamos el coste calculado con el coste almacenado
diferencias = (coste_calculado.round(2) != df["total_cost"].round(2)).sum()

# Mostramos el resultado
print(f"Número de registros con inconsistencias: {diferencias}")
```

***`# ANÁLISIS EXPLORATORIO DE DATOS (EDA)`***


**`## 5.1 Consumo energético por región`**

En este apartado se analiza el consumo energético total de cada región para identificar aquellas con mayor y menor consumo.

```python
# CONSUMO ENERGETICO POR REGION
#=====================================
# 5.1 Consumo energetico por region
# ============================================

# Importamos la librería para crear gráficos
import matplotlib.pyplot as plt

# Calculamos el consumo total por región
consumo_region = df.groupby("region")["consumption_kwh"].sum()

# Mostramos los resultados por pantalla
print(consumo_region)

# Creamos el gráfico de barras
plt.figure(figsize=(8,5))
consumo_region.plot(kind="bar")

# Añadimos un título al gráfico
plt.title("Consumo energético por región")

# Etiquetas de los ejes
plt.xlabel("Región")
plt.ylabel("Consumo total (kWh)")

# Ajustamos la presentación
plt.xticks(rotation=0)

# Mostramos el gráfico
plt.show()
```

**`# 5.2 Consumo energético por tipo de contrato`**

En este apartado se analiza el consumo energético total según el tipo de contrato de los clientes para identificar cuál presenta un mayor consumo.

```python
# CONSUMO ENERGETICO POR TIPO DE CONTRATO
#==============================================
# 5.2 Consumo energético por tipo de contrato
#==============================================

# Importamos la librería para crear gráficos
import matplotlib.pyplot as plt

# Calculamos el consumo total por tipo de contrato
consumo_contrato = df.groupby("contract_type")["consumption_kwh"].sum()

# Mostramos los resultados por pantalla
print(consumo_contrato)

# Creamos el gráfico de barras
plt.figure(figsize=(8,5))
consumo_contrato.plot(kind="bar")

# Añadimos un título al gráfico
plt.title("Consumo energético por tipo de contrato")

# Etiquetas de los ejes
plt.xlabel("Tipo de contrato")
plt.ylabel("Consumo total (kWh)")

# Ajustamos la presentación
plt.xticks(rotation=0)

# Mostramos el gráfico
plt.show()
```

**`# 5.3 Evolución temporal del consumo energético`**

En este apartado se analiza la evolución del consumo energético a lo largo del tiempo agrupando los datos por fecha de facturación.

```python
# EVOLUCION TEMPORAL DEL CONSUMO ENERGETICO
#=====================================================
# 5.3 Evolución temporal del consumo energético
# ====================================================

# Importamos la librería para crear gráficos
import matplotlib.pyplot as plt

# Agrupamos el consumo por fecha de facturación
consumo_fecha = df.groupby("billing_date")["consumption_kwh"].sum()

# Mostramos los resultados
print(consumo_fecha)

# Creamos el gráfico
plt.figure(figsize=(12,5))
consumo_fecha.plot()

# Añadimos el título
plt.title("Evolución temporal del consumo energético")

# Etiquetas de los ejes
plt.xlabel("Fecha de facturación")
plt.ylabel("Consumo total (kWh)")

# Mostramos el gráfico
plt.show()
```

**`# 5.4 Correlación entre consumo, emisiones de CO₂ y coste`**

En este apartado se analiza la relación existente entre el consumo energético, las emisiones de CO₂ y el coste total mediante una matriz de correlación.

```python
# CORRELACION ENTRE CONSUMO, EMISIONES DE CO₂ Y COSTE
#================================================================
# 5.4 Correlación entre consumo, emisiones de CO₂ y coste
# ===============================================================

# Calculamos la matriz de correlación
correlacion = df[[
    "consumption_kwh",
    "co2_emissions",
    "total_cost"
]].corr()

# Mostramos la matriz
print(correlacion)
```

**`# 5.5 Clientes con mayor consumo y mayor coste`**

En este apartado se identifican los clientes con mayor consumo energético y aquellos que generan un mayor coste total.

```python
# CLIENTES CON MAYOR CONSUMO Y MAYOR COSTE
#================================================
# 5.5 Clientes con mayor consumo y mayor coste
#================================================

# Top 10 clientes con mayor consumo
top_consumo = (
    df.groupby("client_name")["consumption_kwh"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

print("Top 10 clientes por consumo energético")
print(top_consumo)

print("\n" + "="*50 + "\n")

# Top 10 clientes con mayor coste
top_coste = (
    df.groupby("client_name")["total_cost"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

print("Top 10 clientes por coste total")
print(top_coste)
```

***`# 6 TABLAS DE CONSUMO`***

**`# 6.1 Preparación de la tabla de consumo por región`**

En este apartado se genera una tabla agregada con el consumo energético total por región.

```python
# PREPARACION DE LA TABLA DE CONSUMO POR REGION
#=======================================================
# 6.1 Tabla de consumo por región
#=======================================================

# Agrupamos el consumo por región
tabla_region = (
    df.groupby("region", as_index=False)["consumption_kwh"]
      .sum()
)

# Renombramos la columna para hacerla más descriptiva
tabla_region.rename(columns={
    "consumption_kwh": "total_consumption_kwh"
}, inplace=True)

# Mostramos la tabla
print(tabla_region)
```

**`# 6.2 Tabla de consumo por tipo de contrato`**

En este apartado se genera una tabla agregada con el consumo energético total según el tipo de contrato.

```python
# TABLA DE CONSUMO POR TIPO DE CONTRATO
#===============================================
# 6.2 Tabla de consumo por tipo de contrato
#===============================================

# Agrupamos el consumo por tipo de contrato
tabla_contrato = (
    df.groupby("contract_type", as_index=False)["consumption_kwh"]
      .sum()
)

# Renombramos la columna para hacerla más descriptiva
tabla_contrato.rename(columns={
    "consumption_kwh": "total_consumption_kwh"
}, inplace=True)

# Mostramos la tabla
print(tabla_contrato)
```

**`# 6.3 Tabla de consumo por mes`**

En este apartado se genera una tabla con el consumo energético total por mes. Esta información permitirá analizar la evolución mensual.

```python
# TABLA DE CONSUMO POR MES
#===================================
# 6.3 Tabla de consumo por mes
# ==================================

# Creamos una nueva columna con el mes de la fecha de facturación
df["month"] = df["billing_date"].dt.to_period("M").astype(str)

# Agrupamos el consumo por mes
tabla_mes = (
    df.groupby("month", as_index=False)["consumption_kwh"]
      .sum()
)

# Renombramos la columna para hacerla más descriptiva
tabla_mes.rename(columns={
    "consumption_kwh": "total_consumption_kwh"
}, inplace=True)

# Mostramos la tabla
print(tabla_mes)
```

**`# 6.4 Cálculo de los indicadores clave (KPIs)`**

En este apartado se calculan los principales indicadores del conjunto de datos, que posteriormente se utilizarán como tarjetas informativas.

```python
# CALCULO DE LOS KPIs
#====================================
# 6.4 Cálculo de los KPIs
#====================================

# Consumo total
consumo_total = df["consumption_kwh"].sum()

# Coste total
coste_total = df["total_cost"].sum()

# Emisiones totales de CO₂
co2_total = df["co2_emissions"].sum()

# Porcentaje medio de energía renovable
renovable_media = df["renewable_energy_percentage"].mean()

# Mostrar resultados
print("Consumo total (kWh):", consumo_total)
print("Coste total (€):", round(coste_total, 2))
print("Emisiones totales CO₂:", round(co2_total, 2))
print("Porcentaje medio de energía renovable:", round(renovable_media, 2), "%")
```

**`# 6.5 Exportación de las tablas para Looker Studio`**

En este apartado se exportan las tablas generadas durante el análisis a archivos CSV, que posteriormente serán utilizados como fuente de datos en Looker Studio.

```python
# EXPORTACION DE LAS TABLAS
#======================================
# 6.5 Exportación de las tablas
#======================================

# Exportamos las tablas a archivos CSV

tabla_region.to_csv("tabla_region.csv", index=False)

tabla_contrato.to_csv("tabla_contrato.csv", index=False)

tabla_mes.to_csv("tabla_mes.csv", index=False)

# Mostramos un mensaje de confirmación

print("Las tablas se han exportado correctamente.")
print("Archivos generados:")
print("- tabla_region.csv")
print("- tabla_contrato.csv")
print("- tabla_mes.csv")
```

**`# 6.6 Tabla de indicadores KPI`**

En este apartado se generan los principales indicadores (KPI) que se utilizarán en el panel de Google Looker Studio.

Se calculan métricas globales del conjunto de datos como el consumo energético total, consumo medio, emisiones totales de CO₂, coste total, coste medio por kWh y porcentaje medio de energía renovable.

Finalmente, la tabla se exporta a un archivo CSV para facilitar su integración en el dashboard.

```python
# TABLA INDICADORES KPIS
# ====================================================
# 6.6 Tabla indicadores Kpis para Looker Studio
# ====================================================

# Creamos una tabla con las métricas principales
tabla_kpis = pd.DataFrame({
    "consumo_total_kwh": [df["consumption_kwh"].sum()],
    "consumo_promedio_kwh": [df["consumption_kwh"].mean()],
    "emisiones_totales_co2": [df["co2_emissions"].sum()],
    "coste_total": [df["total_cost"].sum()],
    "coste_promedio_kwh": [df["cost_per_kwh"].mean()],
    "porcentaje_renovable_medio": [df["renewable_energy_percentage"].mean()]
})

# Mostramos la tabla generada
display(tabla_kpis)

# Exportamos la tabla a un archivo CSV
tabla_kpis.to_csv("tabla_kpis.csv", index=False)

# Mensaje de confirmación
print("La tabla KPI se ha exportado correctamente.")
print("Archivo generado:")
print("- tabla_kpis.csv")
```
